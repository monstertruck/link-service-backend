```python
import httpx
import json

BASE_URL = "http://localhost:8000"

def pretty(response: httpx.Response) -> None:
    print(f"Status: {response.status_code}")
    try:
        print(json.dumps(response.json(), indent=2))
    except Exception:
        print(response.text)
```

## POST /links — submit a new URL (auto-summarize + categorize)

```python
sample_link = "https://om.co/2026/05/15/ai-in-everything-everywhere/"

response = httpx.post(f"{BASE_URL}/links", json={
    "url": sample_link,
})
pretty(response)
```

## POST /links — with a pre-supplied summary (skips Claude summarization)

```python
response = httpx.post(f"{BASE_URL}/links", json={
    "url": "https://news.ycombinator.com",
    "title": "Hacker News",
    "summary": "Hacker News is a social news site focused on computer science and entrepreneurship.",
})
pretty(response)
```

## POST /links — duplicate URL (should return 409 with existing record)

```python
response = httpx.post(f"{BASE_URL}/links", json={
    "url": "https://news.ycombinator.com",
})
pretty(response)
```

## GET /links — list all links

```python
response = httpx.get(f"{BASE_URL}/links")
pretty(response)
```

## GET /links — filter by category

```python
response = httpx.get(f"{BASE_URL}/links", params={"category": "tech"})
pretty(response)
```

## GET /links — filter by status

```python
response = httpx.get(f"{BASE_URL}/links", params={"status": "unread"})
pretty(response)
```

## GET /links/{id} — fetch a single link

```python
LINK_ID = 1  # change to a valid ID from above

response = httpx.get(f"{BASE_URL}/links/{LINK_ID}")
pretty(response)
```

## PATCH /links/{id}/status — mark as read

```python
response = httpx.patch(f"{BASE_URL}/links/{LINK_ID}/status", params={"status": "read"})
pretty(response)
```

## GET /categories — all categories with link counts

```python
response = httpx.get(f"{BASE_URL}/categories")
pretty(response)
```

## DELETE /links/{id}

```python
response = httpx.delete(f"{BASE_URL}/links/{LINK_ID}")
print(f"Status: {response.status_code}")  # 204 = deleted, 404 = not found
```
